# Tutorial on Scraping NFL Game Data

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import random
import time
```

```python
# This creates a list of strings holding all seasons between 2014 and 2023
seasons = [str(season) for season in range(2014, 2023)]

print(f'number of seasons={len(seasons)}')
```

```python
# This craetes a list of the 3 letter abbreviations for the NFL teams you may see in these years
teams = ['crd', 'atl', 'rav', 'buf', 'car', 'chi', 'cin', 'cle', 'dal', 'den', 'det',
         'gnb', 'htx', 'clt', 'jax', 'kan', 'sdg', 'ram', 'rai', 'mia', 'min', 'nwe',
         'nor', 'nyg', 'nyj', 'phi', 'pit', 'sea', 'sfo', 'tam', 'oti', 'was']
print(f'number of teams={len(teams)}')
```

```python
# Confirm we are running the virtual environment set up within our NFLPlayerStats folder
import sys
print(sys.executable)
```

```python
# Create the dataframe to which we will be appending data to
NflDf = pd.DataFrame()

# This takes our list of seasons and iterates through them
for season  in seasons:
    # We want for each team all games within that season
    for team in teams:
        # We are setting the URL here with both of our variables we iterate on
        url = 'https://www.pro-football-reference.com/teams/' + team + '/' + season + '/gamelog'
        print(url)

        # Get offensive stats for each team
        OffDf = pd.read_html(url, header = 1, attrs = {'id': 'gamelog' + season})[0]

        # Get defensive stats for each team
        DefDf = pd.read_html(url, header = 1, attrs = {'id': 'gamelog_opp' + season})[0]

        # Concatenate both offense and defense dataframes
        TeamDf = pd.concat([OffDf, DefDf], axis = 1)

        # Insert the season as a column of our dataframe
        TeamDf.insert(loc = 0, column = 'Season', value = season)

        # Insert the team as a column of our dataframe in the third column. Upper case abbreviations
        TeamDf.insert(loc = 2, column = 'Team', value = team.upper())

        # Here we put a sleep timer for our html requests. At this time it is limiting to 10
        # Requests per minute (1/21/2025)
        time.sleep(random.randint(7,8))

    # Display the dataframe
    print(TeamDf)

```

```python
# Set dataframe
NflDf = TeamDf
```

```python
# Save NFL data as a CSV File
NflDf.to_csv('NFLGamelogs2014To2022.csv')

print(NflDf.shape)
```
